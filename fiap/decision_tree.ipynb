# Imports

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
```

# Simulating e-mail data

```python
emailData = pd.DataFrame({
    'suspicious_words': [1, 0, 0, 1, 1, 0], # 1 true, 0 false
    'trustworthy_domain': [0, 1, 1, 0, 0, 1], # 1 true, 0 false
    'classification': [1, 0, 0, 1, 1, 0] # 1 phishing, 0 safe
})
```

# Separating characteristics and target

```python
x = emailData[['suspicious_words', 'trustworthy_domain']]
y = emailData['classification']
```

# Creating Decision Tree Model

```python
model = DecisionTreeClassifier()
model.fit(x, y)
```

# Visualizing Decision Tree

```python
plt.figure(figsize=(10,8))
tree.plot_tree(model, filled=True, rounded=True,
               feature_names=['suspicious_words', 'trustworthy_domain'],
               class_names=['Safe', 'Phishing'])
plt.show()
```
